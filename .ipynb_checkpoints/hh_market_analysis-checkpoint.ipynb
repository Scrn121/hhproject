```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy import create_engine
import warnings
warnings.filterwarnings('ignore')

plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
%matplotlib inline

import os
from dotenv import load_dotenv

load_dotenv()

DB_USER = os.getenv('POSTGRES_USER')
DB_PASS = os.getenv('POSTGRES_PASSWORD')
DB_HOST = 'localhost'
DB_PORT = os.getenv('POSTGRES_PORT')
DB_NAME = os.getenv('POSTGRES_DB')

engine = create_engine(
    f'postgresql://{DB_USER}:{DB_PASS}@{DB_HOST}:{DB_PORT}/{DB_NAME}'
)

print('Sucsess')
```

```python
salary_by_skill = pd.read_sql("select * from marts.mart_salary_by_skill", engine)
top_employers = pd.read_sql("select * from marts.mart_top_employers", engine)
vacancy_trends = pd.read_sql("select * from marts.mart_vacancy_trends", engine)

vacancies = pd.read_sql("""
    select
        vacancy_name,
        salary_from,
        salary_to,
        salary_avg,
        employer_name,
        area_name,
        requirement,
        published_date
    from staging.stg_vacancies
    where salary_from > 0
""", engine)

print(f"Vacancies:{len(vacancies)}")
print(f"skills: {len(salary_by_skill)}")
print(f"Employers: {len(top_employers)}")
```

```python
#salaries
plt.figure(figsize=(12,6))

plt.subplot(1, 2, 1)
plt.hist(vacancies['salary_avg'], bins=30, edgecolor='black', alpha=0.7)
plt.xlabel('Средняя зарплата (₽)')
plt.ylabel('Количество вакансий')
plt.title('Распределение зарплат')
plt.axvline(vacancies['salary_avg'].median(), color='red', linestyle='--', 
            label=f'Медиана: {vacancies["salary_avg"].median():.0f}₽')
plt.legend()

plt.subplot(1, 2, 2)
plt.boxplot(vacancies['salary_avg'])
plt.ylabel('Зарплата (₽)')
plt.title('Boxplot: выбросы в зарплатах')

plt.tight_layout()
plt.show()

print(f"average_salary: {vacancies['salary_avg'].mean():.0f}₽")
print(f"median_salary: {vacancies['salary_avg'].median():.0f}₽")
print(f"st_deviation: {vacancies['salary_avg'].std():.0f}₽")
```

```python
#top 10 skills
plt.figure(figsize=(10, 6))

top_skills = salary_by_skill.nlargest(10, 'avg_salary')

plt.barh(top_skills['skill'], top_skills['avg_salary'], color='steelblue')
plt.xlabel('Средняя зарплата (₽)')
plt.title('Топ-10 навыков по зарплате')
plt.gca().invert_yaxis()  # Самый высокий навык сверху

for i, v in enumerate(top_skills['avg_salary']):
    plt.text(v + 5000, i, f'{v:,.0f}₽', va='center')

plt.tight_layout()
plt.show()
```

```python
# Top 10 employers
plt.figure(figsize=(10, 6))

top_10_employers = top_employers.nlargest(10, 'vacancy_count')

plt.barh(top_10_employers['employer_name'], top_10_employers['vacancy_count'], 
         color='green')
plt.xlabel('Количество вакансий')
plt.title('Топ-10 работодателей по активности')
plt.gca().invert_yaxis()

plt.tight_layout()
plt.show()
```

```python
# Vacancies in cities
city_salaries = vacancies.groupby('area_name')['salary_avg'].agg(['mean', 'count'])
city_salaries = city_salaries[city_salaries['count'] >= 3]  # Минимум 3 вакансии
city_salaries = city_salaries.sort_values('mean', ascending=False)

plt.figure(figsize=(10, 6))
plt.bar(city_salaries.index, city_salaries['mean'], color='lightgreen')
plt.xlabel('Город')
plt.ylabel('Средняя зарплата (₽)')
plt.title('Средняя зарплата по городам')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
